```python
import os
import pandas as pd
import numpy as np
import tensorflow as tf
import mlflow
import mlflow.tensorflow
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from codecarbon import EmissionsTracker
import matplotlib.pyplot as plt
```

```python
mlflow.set_tracking_uri("file://" + os.path.abspath("../mlruns"))
print("Tracking URI:", mlflow.get_tracking_uri())
```

```python
df = pd.read_csv("../data/processed/merged_queries.csv")
```

```python
# Prepare features and labels
X = df[[f"f{i}" for i in range(768)]].values
y = df["label"].values

# Train-validation split
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
# Define model builder function
def build_model(input_dim, dropout_rate):
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(256, activation='relu', input_shape=(input_dim,)),
        tf.keras.layers.Dropout(dropout_rate),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Start emissions tracking
tracker = EmissionsTracker(output_dir="./", log_level="error")
tracker.start()

mlflow.set_experiment("milestone4_training")

# Hyperparameter configurations
hyperparams_list = [
    {"dropout_rate": 0.3, "batch_size": 32, "epochs": 10},
    {"dropout_rate": 0.5, "batch_size": 64, "epochs": 15},
    {"dropout_rate": 0.2, "batch_size": 128, "epochs": 20},
]

for idx, params in enumerate(hyperparams_list):
    with mlflow.start_run(run_name=f"Run_{idx+1}"):
        mlflow.tensorflow.autolog()
        mlflow.log_params(params)

        model = build_model(input_dim=768, dropout_rate=params["dropout_rate"])
        history = model.fit(
            X_train, y_train,
            validation_data=(X_val, y_val),
            epochs=params["epochs"],
            batch_size=params["batch_size"],
            verbose=1
        )

        # Final evaluation
        from sklearn.metrics import classification_report, roc_auc_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay

        y_probs = model.predict(X_val).flatten()
        y_pred = (y_probs > 0.5).astype(int)

        #  FIX: Evaluate to get `acc`
        loss, acc = model.evaluate(X_val, y_val, verbose=0)

        # Metrics (macro for multiclass-safe)
        mlflow.log_metric("final_val_accuracy", acc)
        mlflow.log_metric("precision_macro", precision_score(y_val, y_pred, average='macro'))
        mlflow.log_metric("recall_macro", recall_score(y_val, y_pred, average='macro'))
        mlflow.log_metric("f1_macro", f1_score(y_val, y_pred, average='macro'))

        try:
            mlflow.log_metric("roc_auc", roc_auc_score(y_val, y_probs))
        except:
            print("ROC AUC not applicable (likely due to multiclass)")

        # Save classification report
        report_text = classification_report(y_val, y_pred, output_dict=False)
        print("Classification Report:\n", report_text)

        with open("classification_report.txt", "w") as f:
            f.write(report_text)
        mlflow.log_artifact("classification_report.txt")

        # Confusion matrix
        cm = confusion_matrix(y_val, y_pred)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm)
        disp.plot()
        plt.title(f"Confusion Matrix - Run {idx+1}")
        plt.show()

        # Accuracy plot
        plt.figure(figsize=(8, 4))
        plt.plot(history.history['accuracy'], label='Train Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy")
        plt.title(f"Training vs Validation Accuracy - Run {idx+1}")
        plt.grid(True)
        plt.legend()
        plt.show()

# Stop emissions tracker
tracker.stop()
```

```python
import os
import time
import mlflow
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.metrics.pairwise import cosine_similarity
```

```python
mlflow.set_tracking_uri("file://" + os.path.abspath("../mlruns"))
print("Tracking URI:", mlflow.get_tracking_uri())
mlflow.set_experiment("milestone4_query_evaluation")
```

```python
# Simulate a few 100-dimensional query embeddings (normally from CLIP model)
query_embeddings = {
    "red t-shirt": np.random.rand(1, 700),
    "laptop": np.random.rand(1, 700),
    "keyboard": np.random.rand(1, 700),
    "smart watch": np.random.rand(1, 700),
}
```

```python
# Load product features from training dataset (100 features used)
product_df = pd.read_csv("../data/processed/small_input.csv")
product_vectors = product_df[[f"f{i}" for i in range(700)]].values
```

```python
# Evaluate and log metrics for each query into MLflow
for query, q_embed in query_embeddings.items():
    start_time = time.perf_counter()

    # Cosine similarity with all product vectors
    sims = cosine_similarity(q_embed, product_vectors)[0]
    top_k = sims.argsort()[-3:][::-1]
    top_scores = sims[top_k]
    
    duration_sec = time.perf_counter() - start_time
    duration_ms = duration_sec * 1000
    
    avg_sim = np.mean(top_scores)

    print(f"Query: {query}")
    print(f"Top-3 Cosine Similarities: {top_scores}")
    print(f"Response time: {duration_sec:.4f} sec ({duration_ms:.2f} ms)")

    # Log to MLflow
    with mlflow.start_run(run_name=f"query: {query}"):
        mlflow.log_param("query_text", query)
        mlflow.log_metric("avgerage cosine similarity", float(avg_sim))
        mlflow.log_metric("response time in sec", duration_sec)
        mlflow.log_metric("response time in ms", duration_ms)
        for i, score in enumerate(top_scores):
            mlflow.log_metric(f"top_{i+1}_similarity", float(score))
```
